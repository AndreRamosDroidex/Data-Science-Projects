# top 10 most popular artists on tiktok in 2020

```python
plt.style.use("bmh")
plt.figure(figsize=(15,10))
popular_songs["artist_name"].value_counts().head(10).plot(kind="barh",title="10 most popular artists on TikTok in 2020")
```

```python
plt.style.use("bmh")
plt.figure(figsize=(15,10))
popular_songs["artist_name"].value_counts().head(10).plot(kind="pie",title="10 most popular artists on TikTok in 2020",ylabel="")
```
